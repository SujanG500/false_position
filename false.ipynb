```python
import matplotlib.pyplot as plt
import numpy as np

def false_position_method(func, a, b, tol=1e-6, max_iter=100):
    if func(a) * func(b) >= 0:
        raise ValueError("Function values at interval endpoints must have opposite signs.")
    
    iter_count = 0
    approx_roots = []
    
    while iter_count < max_iter:
        c = (a * func(b) - b * func(a)) / (func(b) - func(a))
        approx_roots.append(c)
        
        if abs(func(c)) < tol:
            return c, approx_roots
        
        if func(c) * func(a) < 0:
            b = c
        else:
            a = c
        
        iter_count += 1
    
    raise ValueError(f"Method did not converge after {max_iter} iterations.")

# Example usage
def func(x):
    return x - 2**-x

a = 0
b = 1

# Visualization setup
x_vals = np.linspace(a, b, 400)
y_vals = func(x_vals)
root, approx_roots = false_position_method(func, a, b)

plt.figure(figsize=(10, 6))

# Plot the function
plt.plot(x_vals, y_vals, label='Function')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--', label='y = 0')

# Iterative process visualization
for i, approx_root in enumerate(approx_roots):
    plt.scatter(approx_root, func(approx_root), color='red', marker='o', label=f'Iteration {i + 1}')
    plt.plot([a, b], [func(a), func(b)], color='green', linestyle='dashed')
    plt.legend()
    plt.title('False Position Method')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.show()

print("Approximate root:", root)
print("Function value at root:", func(root))
```
